```python
import math 

def f(x):
	return math.sin(x**2/2)/math.log(x+4,2)


def search(ind, d):
	if ind+d > 10:
		if f(ind) > f(ind-d):
			return ind
		else:
			return search(ind-d,d)

	elif ind - d < 0 :
		if f(ind) > f(ind+d) :
			return ind
		else:
			return search(ind+d, d)

	else : 
		if f(ind) > f(ind+d) and f(ind) > f(ind - d):
			return ind

		elif f(ind+d) >= f(ind-d):
			return search(ind+d,d)

		else:
			return search(ind-d,d)

res = [[0 for i in range(11)] for j in range(10)] 
delta = 0.01

for i in range(10):
	for j in range(11):
		res[i][j] = search(j,delta)

	delta = delta + 0.01

for i in range(10):
	print(res[i])
	
	
```
